```python
import json
import pandas as pd
```

```python
with open('raw_results.json') as f:
  data = json.load(f)
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# 33

df['league_id'] = df['team_id'].apply(lambda i: 1 if int(i) < 34 else 2)
```

```python
df.info()
```

```python
df.head()
```

```python
numeric_cols = ['id','game_season', 'team_id', 'opponent_id', 'team_points', 'opponent_points']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, downcast='integer', axis=1)
```

```python
df.info()
```

```python
df['game_date'] = pd.to_datetime(df['game_date'])
```

```python
df.info()
```

```python
bool_dict = {'True': 1, 'False': 0}

df['is_team_home'].replace(bool_dict, inplace=True)
df['is_overtime'].replace(bool_dict, inplace=True)

df.info()
```

```python
df.head()
```

```python
df[df['league_id'] == 2]
```

```python
def game_identifier(game):
    date_identifier = str(game['game_date'].date())
    
    team_identifier = "-".join(sorted([str(game['team_id']), str(game['opponent_id'])]))
    
    points_identifier = str(game['team_points'] + game['opponent_points'])
    
    return "-".join([date_identifier, team_identifier, points_identifier])

df['game_identifier'] = df.apply(game_identifier, axis=1)

df.head()
```

```python
deduped_df = df.drop_duplicates(subset='game_identifier')
```

```python
df.head()
```

```python
df['win'] = df.apply(lambda game: 1 if game['team_points'] > game['opponent_points'] else 0, axis=1)
```

```python
df.head()
```

```python
# df['loss'] = df.apply(lambda game: 1 if game['win'] == 0 and game['is_overtime'] == 0 else 0)

df['loss'] = df.apply(lambda game: 1 if (game['team_points'] < game['opponent_points']) and game['is_overtime'] == 0 else 0, axis=1)
```

```python
df.head()
```

```python
df['ot_loss'] = df.apply(lambda game: 1 if (game['team_points'] < game['opponent_points']) and game['is_overtime'] == 1 else 0, axis=1)
```

```python
df.head(25)
```

```python
season_ungrouped_df = df[['league_id', 'team_id', 'game_season', 'win', 'loss', 'ot_loss']]
```

```python
season_ungrouped_df.tail()
```

```python
season_grouped_df = season_ungrouped_df.groupby(by=['league_id', 'team_id', 'game_season']).sum()
```

```python
season_grouped_df.head()
```

```python
season_grouped_df.tail(25)
```

```python
season_grouped_df.info()
```

```python
season_grouped_df.head()
```

```python
season_grouped_df.reset_index(inplace=True)
```

```python
season_grouped_df.head()
```

```python
nhl_season_df = season_grouped_df[season_grouped_df['league_id'] == 2]
```

```python
len(nhl_season_df)
```

```python
nhl_season_df.head()
```

```python
def calculate_points(season):
    return (2 * season['win']) + season['ot_loss']

nhl_season_df['points'] = nhl_season_df.apply(calculate_points, axis=1)
```

```python
nhl_season_df.tail(9)
```

```python
subset_df = nhl_season_df[['team_id', 'game_season', 'points']]
subset_df.plot(x='game_season', y='points', figsize=(20,10))
```
